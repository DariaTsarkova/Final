```python
import pandas as pd
```

```python
movies_stats = pd.read_csv(r'C:\Users\Dasha\Final\movies_stats.csv') # загружаем файл
movies_stats.head()
```

```python
years = list(range(1950, 2011)) # в переменную years записываем список из всех годов с 1950-го по 2010-й
years[:10]
```

```python
def production_year(title): # функция, которая каждой строке из названия фильма выставляет год выпуска (год выпуска берём из столбца title).
    for year in years:
        if str(year) in title:
            return year
    return 1900 # если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год
```

```python
movies_stats['year'] = movies_stats['title'].apply(production_year) # записываем год выпуска фильма в новый столбец ‘year’
movies_stats.head()
```

```python
movies_stats_result = movies_stats.groupby('year').agg({'rating' : 'mean'}) # посчитаем средний рейтинг всех фильмов для каждого значения столбца ‘year’
movies_stats_result
```

```python
movies_stats_result_sorted = movies_stats_result.sort_values(by='rating', ascending=False) # отсортируем результат по убыванию рейтинга
print(movies_stats_result_sorted.head(60))
```
